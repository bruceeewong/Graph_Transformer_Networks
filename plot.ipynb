```python
from utils import load_attention_weights, plot_attention_weights, print_attention_weights, get_edge_type_names, print_top_metapaths
```

```python
def identify_db_name(filename):
    last_comp =  filename.split('/')[-1]
    basename = last_comp.split('.')[0]
    return basename.split('_')[-1]

def analyze_attention_weights(filename, num_channels=1):
    # Load saved data (no need to retrain)

    attn_data = load_attention_weights(filename)
    db_name = identify_db_name(filename)
    edge_type_names = get_edge_type_names(db_name)
    print(f"Analyzing {db_name} dataset, edge types: {edge_type_names}")
    # Tweak visualization
    for ch in range(num_channels):
        plot_attention_weights(attn_data, 
                            edge_type_names=edge_type_names,
                            channel=ch)
    print_attention_weights(attn_data, edge_type_names=edge_type_names)
    print_top_metapaths(attn_data, dataset=db_name, top_k=3, connected_only=True)
```

```python
for filename in [
    './output/20251208_033739/attention_weights_DBLP.npz',
]:
    analyze_attention_weights(filename, num_channels=2)
```
